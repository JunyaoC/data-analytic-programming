### Exercise  1
Given a data set Melbourne Housing Dataset. This dataset contains 21 columns. The details of the specific variables is as follows:

<b>Rooms:</b> no. of rooms<br>
<b>Price:</b> Price in dollars<br>
<b>Method:</b> S - property sold; SP - property sold prior; PI - property passed in; PN - sold prior not disclosed; SN - sold not disclosed; NB - no bid; VB - vendor bid; W - withdrawn prior to auction; SA - sold after auction; SS - sold after auction price not disclosed. N/A - price or highest bid not available.
<b>Type:</b> br - bedroom(s); h - house,cottage,villa, semi,terrace; u - unit, duplex; t - townhouse; dev site - development site; o res - other residential.<br>
<b>SellerG:</b> Real Estate Agent<br>
<b>Date:</b> Date sold<br>
<b>Distance:</b> Distance from CBD<br>
<b>Regionname:</b> General Region (West, North West, North, North east …etc)<br>
<b>Propertycount:</b> Number of properties that exist in the suburb.<br>
<b>Bedroom2 :</b> Scraped # of Bedrooms (from different source)<br>
<b>Bathroom:</b> Number of Bathrooms<br>
<b>Car:</b> Number of carspots<br>
<b>Landsize:</b> Land Size<br>
<b>BuildingArea:</b> Building Size<br>
<b>CouncilArea:</b> Governing council for the area<br>

detail: https://www.kaggle.com/dansbecker/melbourne-housing-snapshot/home?select=melb_data.csv


1. Read the dataset using the `read_csv` function in pandas

```python
import pandas as pd
data = pd.read_csv('melb_data.csv')
data.head()
```

2. Using the knowledge you learnt in chapter 5, check each column if there is any column with missing values

```python
columnMissingDF = []

for column in data:
    x = data[data[column].isnull()]
    if( len(x) > 0 ):
        columnMissingDF.append({
            'column':column,
            'missing count':len(x)
        })

display(pd.DataFrame.from_dict(columnMissingDF))
```

3. Remove the data entry (rows) that have at least one missing data.

```python
cleansed = data.dropna()

display(cleansed)
```

4. Create a new data frame using multiindex formed using 'Regionname' and 'Type' column

```python
df1 = data
df1 = df1.set_index(["Regionname","Type"])
display(df1)
```

5. Using the new data frame from 4. List the house price for the townhouses (type: t) in Western Metropolitan. <br>Then show the amount of townhouses and average price of the townhouses there.

```python
left = df1.loc[['Western Metropolitan']]
right = df1.swaplevel('Regionname','Type').loc[['t']]
joined = pd.merge(left,right, how="inner")
display(joined)
```

### Exercise 2
Given two sets of data about restaurant visit in Japan:<br>
<b>air_store_info.csv</b> - store's information from AirREGI / Restaurant Board similar to Square.<br> 
<b>air_visit_data.csv</b> - the visit data information to each of the store (refer store ID), date and amount of visitors.<br>
<b>date_info.csv</b>      - the information of the date and holiday flag (if the day is holiday, flag = 1)<br>

The dataset is part of the datasets shared in: https://www.kaggle.com/c/recruit-restaurant-visitor-forecasting/data

1. Read all the datasets separately using the `read_csv` function in pandas

```python
store_info = pd.read_csv('air_store_info.csv')
visit_data = pd.read_csv('air_visit_data.csv')
date_info = pd.read_csv('date_info.csv')
```

2. Create a new data frame by merging the data from `air_visit_data.csv` to `air_store_info.csv` to get the corresponding information regarding the store.

```python
df1 = pd.merge(store_info,visit_data, on=["air_store_id","air_store_id"])
display(df1)
```

3. Using the resulting data frame from (2), merge with `date_info.csv` based on the key `date` into a new data frame that consists of the holiday flag information. 

```python
df2 = pd.merge(df1,date_info, left_on="visit_date", right_on="calendar_date")
display(df2)
```

4. Check the new data frame from (3) if there is any missing values. If yes, remove the data entries (rows) with missing values

```python
columnMissingDF = []
for column in df2:
    x = df2[df2[column].isnull()]
    columnMissingDF.append({
        'column':column,
        'missing count':len(x)
    })
       

display(pd.DataFrame.from_dict(columnMissingDF))
```
