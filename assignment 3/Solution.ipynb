```python
df1.loc[:,'No_Times_Pregnant'] = df1.loc[:,'No_Times_Pregnant'].fillna(1)
```

```python
df1_cleansed1 = df1.dropna(how='any',axis=0,thresh=4)

print(df1_cleansed1.size)
```

```python
df1_cleansed2 = df1_cleansed1.copy()

df1_cleansed2.loc[:,'Plasma_Glucose'] = df1_cleansed2.loc[:,'Plasma_Glucose'].fillna( df1_cleansed2.loc[:,'Plasma_Glucose'].mean())
df1_cleansed2.loc[:,'Diastolic'] = df1_cleansed2.loc[:,'Diastolic'].ffill()
df1_cleansed2.loc[:,'Triceps'] = df1_cleansed2.loc[:,'Triceps'].bfill()
df1_cleansed2.loc[:,'Insulin'] = df1_cleansed2.loc[:,'Insulin'].fillna( df1_cleansed2.loc[:,'Insulin'].mean())
df1_cleansed2.loc[:,'BMI'] = df1_cleansed2.loc[:,'BMI'].fillna( df1_cleansed2.loc[:,'BMI'].median())


# df1_cleansed2.loc[81:,'Diastolic'].fillna(method="ffill")

# df1_cleansed2.loc[:,'Triceps'] = df1_cleansed2.loc[:,'Triceps'].bfill()
# df1_cleansed2.loc[:'Triceps'] = df1_cleansed2.loc[:'Triceps'].bfill()
# df1_cleansed2['Insulin'].fillna( df1_cleansed2['Insulin'].mean())
# df1_cleansed2['BMI'].fillna( df1_cleansed2['BMI'].median() )

df1_cleansed2
```
