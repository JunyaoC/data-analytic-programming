## Numpy

### tasks
1.	Read the “Attendance_RAW.txt” and save the data into a NumPy array.
2.	Shape the NumPy array to a structure where row represents the employees, column represents the working days.
3.	Construct a function that can calculate and return the list of average of attendance for each day.


```python
import numpy as np

def read_and_parse():
    with open('Attendance_RAW.txt') as f:      ## task1 start
        content = f.readlines()
    content = [int(x.strip()) for x in content]

    
    array = np.array(content)                 ## task1 end
    array = np.reshape(array,(14,15))         ## task2 start
    array = np.transpose(array)               ## task2 end
    return array
    
def get_day_avg(array):
#     print(array)
    avg = array.mean(axis=0)
    return avg

array = read_and_parse()
avg = get_day_avg(array)

print(array)
print(avg)
```

```python
def get_avg_department(df):
    
    departments = df['Department'].unique()
    
    attdn = {}
    
    for dp in departments:
        dpAtt = df.loc[df['Department'] == dp]
        attdn[dp] = dpAtt['total'].sum() / len(dpAtt['total'])
    
    return attdn


avgAttdn = get_avg_department(emp_attendance)
print(avgAttdn)
```
