(c)	Read data from Sales Data.csv and show all product sales using a multiline plot, sized 1200 × 600 pixels. Differentiate all the lines by line color and marker.

```python
def randomColor(count):
    result = []
    for i in range(0,count):
        r = np.random.rand()
        b = np.random.rand()
        g = np.random.rand()
        
        result.append( (r, g, b) )
    return result
```

```python
def randomMarker(count):
    markerChoice = [".","o","+",",","x","D","8","s","p","*"]
    return np.random.choice(markerChoice, count, replace=False)
```

```python
productDF = sales.drop(columns=['month_number','total_units','total_profit'])
x = sales['month_number']

plt.figure(figsize=(1200*px,600*px))
plt.xlabel('Month Number')
plt.ylabel('Sold Unit number')
plt.title('Product Sales Per Month')
plt.xticks(x)

colors = randomColor(len(productDF.columns))
markers = randomMarker(len(productDF.columns))

for idx,column in enumerate(productDF.columns):
#     print(column)
    y = productDF[column]
    plt.plot(x, y, linewidth=3,color=colors[idx],marker=markers[idx],ms=8)

plt.legend(productDF.columns, loc ="lower right") 
plt.show()
```

```python
categoryADF = productDF[["facecream","facewash","moisturizer"]]
categoryBDF = productDF[["toothpaste"]]
categoryCDF = productDF[["bathingsoap","shampoo"]]

categoryProfit = pd.DataFrame({'Category A':categoryADF.sum(axis=1),'Category B':categoryBDF.sum(axis=1),'Category C':categoryCDF.sum(axis=1)})

totalProfit = pd.DataFrame({'total':categoryProfit.sum(axis=1)})

categoryProfitPercentage = categoryProfit.copy()
categoryProfitPercentage['Category A'] = (categoryProfit['Category A'] / totalProfit['total']) * 100
categoryProfitPercentage['Category B'] = (categoryProfit['Category B'] / totalProfit['total']) * 100
categoryProfitPercentage['Category C'] = (categoryProfit['Category C'] / totalProfit['total']) * 100

# categoryProfitPercentage['sum'] = categoryProfitPercentage.sum(axis=1)

# categoryProfitPercentage
```

# Task 2

1.	Read the three datasets given below and save the data into different DataFrame.

```python
salesDF = pd.read_csv("sales.csv")
customerDF = pd.read_csv("customer.csv")
productsDF = pd.read_csv("products.csv")
```

Before we continue, let's join the data into one DF

```python
mergedDF = salesDF.merge(customerDF, left_on="c_id",right_on="c_id")
mergedDF = mergedDF.drop(columns=['product'])
mergedDF = mergedDF.merge(productsDF, left_on="p_id",right_on="p_id")
mergedDF['sale_total'] = mergedDF['qty'] * mergedDF['price']
mergedDF = mergedDF.drop(columns=['c_id','p_id'])

mergedDF
```

2.	From the above datasets, produce a chart (any appropriate chart) for each below:
- products sold by each store
- quantity sold against each product
- quantity and total sales against each product


```python
storeProductSaleDF = mergedDF.groupby(["store","product"]).sum()
storeProductSaleDF = storeProductSaleDF.unstack()
storeProductSaleDF = storeProductSaleDF['qty']
storeProductSaleDF = storeProductSaleDF.fillna(0)
```

```python
productQty = mergedDF.groupby(["product"]).sum()
productQty = productQty.unstack()
productQty = productQty['qty']
productQty = productQty.fillna(0)
```

```python
productSaleQty = mergedDF.groupby(["product"]).sum()
productSaleQty = productSaleQty[ ["qty","sale_total"]]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

storeProductSaleDF.plot.bar(ax=ax1)

productQty.plot.bar(ax=ax2)

productSaleQty['qty'].plot.bar(ax=ax3)
productSaleQty['sale_total'].plot.line(ax=ax3,color="red",secondary_y=True)

plt.show()
```
