In pandas library, you can read CSV data file using the ```read_csv``` function in pandas. The function will read the csv data file straight into data frame, with the default setting where 1st row of the data file will become the header of the data frame.


view on github: https://go.midknight.co/e9r

````python
import pandas as pd

data = pd.read_csv('telco-customer-churn.csv')
data.head()
````

````python
# TASK 1

seniorCitizen = data['SeniorCitizen'] == 1
tenureData = data['tenure'] > 12

seniorJoinTenure = pd.merge( data[seniorCitizen],data[tenureData],how='inner')

display(seniorJoinTenure)
````

````python
# TASK 2

churn = data['Churn'].value_counts()
churnCount = churn['Yes']
print(churnCount)

churnDF = data[ data['Churn'] == 'Yes']

tenureMIN = churnDF['tenure'].min()
print(tenureMIN)

tenureMAX = churnDF['tenure'].max()
print(tenureMAX)

tenureAVG = churnDF['tenure'].mean()
print(tenureAVG)
````

````python
# TASK 3

tenureDF = data[data['tenure'] > 12]

resultDF = tenureDF.loc[ :, ['customerID','Churn','InternetService','StreamingTV','tenure','Contract','MonthlyCharges'] ]

resultDF = resultDF.sort_values(['tenure','MonthlyCharges'] , ascending=[True,False])
display(resultDF)
````
