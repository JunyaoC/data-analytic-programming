## Solutions

```python
def leftSkew(num):
    i = 0 # number of rows
    outputStr = ''
    for i in range(0,num):
        rowStr = ('*' * i) + (' ' * (num - i))
        print(rowStr)
        
def rightSkew(num):
    i = num # number of rows
    outputStr = ''
    for i in range(0,num):
        rowStr = ('*' * (num - i)) + (' ' * 0)
        print(rowStr)
        
def invertedPyramid(num):
    if(num % 2) == 0:
        print('Please use only Odd Numbers to build Pyramid')
    else:
        i = 0 # number of rows
        for i in range(0, ((num//2)+1) ):   ## (num//2)+1 : determines the number of rows.
            rowStr = (' ' * i) +  ('*' * (num - (i*2))) + (' ' * i)
            print(rowStr)

def pyramid(num):
    if(num % 2) == 0:
        print('Please use only Odd Numbers to build Pyramid')
    else:
        i = 0 #number of rows
        numRow = ((num//2)+1) ## (num//2)+1 : determines the number of rows.
        for i in range(0, numRow):  
            rowStr = (' ' * (numRow-i)  ) +  ('*' * (1+(i*2)) ) + (' ' * (numRow - i))
            print(rowStr)
        
```

```python
print('Solution A')
leftSkew(10)
print('\nSolution B\n')
rightSkew(10)
print('\nSolution C\n')
invertedPyramid(9)
print('\nSolution D\n')
pyramid(9)
```
