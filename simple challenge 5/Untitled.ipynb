```python
import pandas as pd
```

```python
csvData = pd.read_csv("telco-customer-churn.csv")
```

```python
def difference(arr):
    return arr.max() - arr.min()
```

```python
grouped = csvData.groupby([csvData['InternetService'],csvData['gender']])
grouped_charge = grouped['MonthlyCharges']
grouped_charge.agg(['min','max',difference])
```

```python
securityDF = csvData['OnlineSecurity'].groupby([csvData['SeniorCitizen'],csvData['OnlineSecurity']])
securityDF.agg(['count'])
```

```python
tenure = csvData.groupby([csvData['Churn'],csvData['InternetService']])
tenure = tenure['tenure']
tenure.agg(['min','max',difference])
```
