<b>TASK 1</b>
1. Read the “Diabetes Dataset.csv” and save the data into a DataFrame.


```python
import pandas as pd
diabetes = pd.read_csv("Diabetes Dataset.csv")
diabetes
```

2.	What is the shape of the DataFrame?

```python
print(diabetes.shape)
```

3.	Get the number of missing values for each column.

```python
diabetes.isnull().sum()
```

4.	Fill in all the missing values in column No_Times_Pregnant with value 1.

```python
diabetes.loc[:,'No_Times_Pregnant'] = diabetes.loc[:,'No_Times_Pregnant'].fillna(1)
diabetes.head()
```

5.	Drop all rows which contains less than 4 observation values.

```python
diabetes = diabetes.dropna(thresh=4)
diabetes
```

6.	Fill in all the missing values in below columns with the given values or methods:
<br>
<img src="https://i.imgur.com/00V1GmF.png">

```python
diabetes = diabetes.copy()
diabetes.loc[:,'Plasma_Glucose'] = diabetes.loc[:,'Plasma_Glucose'].fillna(value=diabetes['Plasma_Glucose'].mean())
diabetes.loc[:,'Diastolic'] = diabetes.loc[:,'Diastolic'].fillna(method='ffill')
diabetes.loc[:,'Triceps']= diabetes.loc[:,'Triceps'].fillna(method='bfill')
diabetes.loc[:,'Insulin'] = diabetes.loc[:,'Insulin'].fillna(value=diabetes['Insulin'].mean())
diabetes.loc[:,'BMI'] = diabetes.loc[:,'BMI'].fillna(value=diabetes['BMI'].median())
diabetes
```

7.	Check back the number of missing values for each column. It should now be 0 for all the columns.

```python
diabetes.isnull().sum()
```

8.	Bin the age into below categories. How many patients with diabetes for each category? 

<img src="https://i.imgur.com/hfxtOyu.png">

```python
bins = [x*10 for x in range(2,10)]
result = pd.cut(diabetes['Age'], bins, right=False)
pd.value_counts(result)
```
