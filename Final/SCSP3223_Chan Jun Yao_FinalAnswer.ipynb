```python
# Write your code here
trainDF = pd.read_csv('FE Dataset_Final/data-q5-1.csv')

trainDF.head()


from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

X = trainDF['distance'].values
X = X.reshape(len(list(X)),1)

y = trainDF['price'].values
y = y.reshape(len(list(y)),1)

model.fit(X, y)

```

#### Question 5(b)(i)

```python
# Write your code here
telcoDF = pd.read_csv('FE Dataset_Final/data-q5-3.csv')
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

X = telcoDF.drop('Churn',axis=1)
y = telcoDF['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

#### Question 5(b)(ii)

```python
# Write your code here
model = GaussianNB()
model.fit(X_train, y_train)
y_model = model.predict(X_test)
```

#### Question 5(b)(iii)

```python
# Write your code here
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_model)
```
