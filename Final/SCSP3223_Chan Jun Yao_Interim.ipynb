```python
# Write your code here
# pd.value_counts(cats)

grade_names = ['E', 'D-', 'D', 'D+','C-','C','C+','B-','B','B+','A-','A','A+']
binnedMarks = pd.cut(marks, bins,labels=grade_names)
binnedMarks
```
